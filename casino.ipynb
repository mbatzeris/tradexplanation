# Μαθηματικη προσδοκια

**Το αποτέλεσμα ενός συγκεκριμένου συμβάντος σε μια τυχαία διαδικασία δεν μπορεί να προβλεφθεί. Ωστόσο, μπορούμε να περιορίσουμε τα πιθανά αποτελέσματα με τη δήλωση πιθανότητας.** 

Ο Pierre Simon Laplace (1749-1827) όρισε την πιθανότητα ως την αναλογία του πλήθους τρόπων που μπορεί το συγκεκριμένο συμβάν να προκύψει, προς το πιθανό πλήθος συμβάντων.

Σε αυτό το σημείο είναι σημαντικό να καταλάβουμε την έννοια των μαθηματικών πιθανοτήτων, γνωστή και ως "πλεονέτημα του παίκτη" (αν είναι υπέρ του παίκτη) ή "πλεονέκτημα του καζίνο" (αν είναι κατά του παίκτη):

**Μαθηματική Πιθανότητα = (1 + A) ∗ P − 1**

*όπου*:

P = Πιθανότητα νίκης.

A = Ποσό που μπορείς να κερδίσεις/Ποσό που μπορείς να χάσεις

Οπότε, αν παίξεις κορώνα γράμματα, με κορώνα να κερδίσεις 2 δολάρια και γράμματα να χάσεις 1 δολάριο, η μαθηματική πιθανότητα για κάθε παρτίδα είναι:

Μαθηματική Πιθανότητα = (1 + 2) ∗ .5 − 1
= 3 ∗ .5 − 1
= 1.5 − 1
= .5

Με άλλα λόγια, μπορείς να κερδίσεις κατά μέσο όρο 50 cents σε κάθε παρτίδα.

    Μέσα σε μία σακούλα υπάρχουν μάρκες με αριθμούς από το 1 έως το 100.
    Οι χρήστες μπορούν να στοιχηματίσουν σε μονούς ή ζυγούς αριθμούς.
    Σε αυτό το παιχνίδι, οι αριθμοί 10 και 11 είναι ιδιαίτεροι. Αν στοιχηματίσουμε σε ζυγούς αριθμούς, το 10 θα μετρήσει ως μονός αριθμός, και αν στοιχηματίσουμε σε μονούς αριθμούς το 11 θα μετρήσει ως ζυγός αριθμός. 
    Αν στοιχηματίσουμε σε ζυγούς αριθμούς και βγει το 10, τότε χάνουμε.
    Αν στοιχηματίσουμε σε μονούς αριθμούς και βγει το 11, τότε χάνουμε.

Αν στοιχηματίσουμε σε μονούς αριθμούς, η πιθανότητα να κερδίσουμε είναι 49/100. Αντίστοιχα, η πιθανότητα να κερδίσει το καζίνο είναι 51/100. Συνεπως, για ένα στοίχημα σε μονό αριθμό, το πλεονέκτημα του καζίνο είναι = 51/100–49/100 = 0.02 = 2%

```python
#@title
import random
import matplotlib.pyplot as plt
```

Place your bet 

```python
#@title
#Place your bet:

#User can choose even or odd number :
choice = input("Θέλεις να στοιχηματίσεις σε Ζυγό αριθμό ή σε Μονό αριθμό \n")

#For even :
if choice=="Ζυγό":
    def pickNote():
        #Get random number between 1-100.
        note = random.randint(1,100)
       
        #Check for our game conditions.
        
        #Notice that 10 isn't considered as even number.
        if note%2!=0 or note==10:
            return False
        elif note%2==0:
            return True

#For odd :        
elif choice=="Μονό":
    def pickNote():
        #Get random number between 1-100.
        note = random.randint(1,100)
        
        #Check for our game conditions.
        
        #Notice that 11 isn't considered as odd number.
        if note%2==0 or note==11:
            return False
        elif note%2==1:
            return True  
```

Main function

```python
#@title
#Main function :
def play(total_money, bet_money, total_plays):

    num_of_plays = []
    money = []
    
    #Start with play number 1
    play = 1
  
    for play in range(total_plays):
        #Win :
        if pickNote():
            #Add the money to our funds
            total_money = total_money + bet_money
            #Append the play number
            num_of_plays.append(play)
            #Append the new fund amount
            money.append(total_money)
        
        #Lose :
        else:
            #Add the money to our funds
            total_money = total_money - bet_money 
            #Append the play number
            num_of_plays.append(play)
            #Append the new fund amount
            money.append(total_money)
    
    #Plot the data :
    plt.ylabel('Χρήματα Παίκτη σε $')
    plt.xlabel('Αριθμός στοιχημάτων')
    plt.rcParams["figure.figsize"] = (20,7)
    plt.plot(num_of_plays,money)

    #Final value after all the iterations :
    final_funds.append(money[-1])
    return(final_funds)
```

```python
#@title Σενάριο 5 στοιχημάτων με 100$ 10 φορές

#Create a list for calculating final funds
final_funds= []

#Run 10 iterations :
for i in range(10):
    ending_fund = play(10000,100,5)

#Print the money the player ends with
print("Αριθμός Στοιχημάτων = 5")
print("Ο Παίκτης ξεκίνησε με $10,000")
print("Ο Παίκτης αποχώρησε με $",str(sum(ending_fund)/len(ending_fund)))
```
